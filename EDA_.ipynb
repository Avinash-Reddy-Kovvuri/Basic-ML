```python
#checking the null values in above dataset
Data.isnull().sum()
```

From this we came across with that some columns has huge number of missing values in it.

The two columns Carcinogenics, medical_waste has all null observations.so it is better to remove the two coumns named Carcinogenics, medical_waste from the training data set for a better prediction column unamed

And the columns named ph, sulfate and Trihalomethanes has also null values and after knowing the value of correlation of the final predicted data with it ,we will conclude whether to remove or not the data from training model. 




```python
Data = Data.drop(['medical_waste','Carcinogenics','Unnamed: 0'],axis=1)
```

Yet Removed unnessary data

```python
Data.hist(edgecolor='black',bins = 25, figsize= (20,20))
plt.show()
```

```python
sns.pairplot(data = Data, hue="Potability", size=3)
plt.show()
```

From these visual Data Analysis we know that the columns named  ph, sulfate and Trihalomethanes have 15%, 24%, 5% has null values over the total values and has an impact on the predicted value Potability, so these aren't dropped from the trained data
